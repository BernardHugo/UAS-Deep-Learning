Melakukan pengecekan pada dataset apakah terdapat data-data yang kosong pada dataset. Hasilnya tidak ada data yang kosong didalam dataset tersebut.

Melihat kelas-kelas yang ada pada atribut "label". Berguna juga untuk mengetahui berapa jumlah kelas yang ada pada atribut "label". Kelas-kelas yang ada pada "label" adalah sadness, anger, love, surprise, fear, dan joy.

Melakukan preprocessing

```python
x = df['text'].values
y = df['label'].values
```

```python
label = {
    'sadness': 0,
    'anger': 1,
    'love': 2,
    'surprise': 3,
    'fear': 4,
    'joy': 5
}

encoded_y = [label[i] for i in df["label"]]

print(encoded_y)
```

Melakukan preprocessing untuk mengubah kategori kelas-kelas yang ada pada atribut "label" dari string menjadi numerik.

**Bagian b**

**Membagi dataset menjadi training, validation, dan test (70:15:15)**

```python
train_size = int(len(x) * 0.7)
test_size = int(len(x) * 0.15)
val_size = len(x) - (train_size - test_size)

x_train, y_train = x[:train_size], encoded_y[:train_size]
x_test, y_test = x[train_size:train_size + test_size], encoded_y[train_size:train_size + test_size]
x_val, y_val = x[train_size + test_size:], encoded_y[train_size + test_size:]
```

**Bagian c**

**Membuat DistilBERT model**

```python
pip install --upgrade keras-nlp
```

```python
import keras_nlp

preprocessor = keras_nlp.models.DistilBertPreprocessor.from_preset(
    "distil_bert_base_en_uncased",
    sequence_length = 130
)

model = keras_nlp.models.DistilBertClassifier.from_preset(
    "distil_bert_base_en_uncased",
    num_classes = 6,
    preprocessor = preprocessor
)
```

```python
model.fit(x = x_train, y = y_train, validation_data = (x_val, y_val), batch_size = 32, epochs = 5)
```

**Bagian d**

**Menemukan skor precision, recall, F1-Score dan accuracy**

```python
from sklearn.metrics import classification_report
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

y_pred = model.predict(x_test)
y_pred = np.argmax(y_pred, axis = 1)

print(classification_report(y_test, y_pred))
```

```python
print(f"Prescision:  {precision_score(y_test, y_pred, average = 'macro') :.3f}")
print(f"Recall: {recall_score(y_test, y_pred, average = 'macro') :.3f}")
print(f"F1-Score: {f1_score(y_test, y_pred, average = 'macro') :.3f}")
print(f"Accuracy: {accuracy_score(y_test, y_pred) :.3f}")
```

Dari hasil precision, recall, F1-Score dan accuracy sangat tinggi yang berarti model DistilBERT yang dibuat melakukan kinerja yang baik dalam melakukan klasifikasi. Nilai precision yang dihasilkan sebesar 0.912 (91.2%), recall sebesar 0.899 (89.9%), F1-Score sebesar 0.902 (90.2%), dan accuracy sebesar 0.932 (93.2%).

Nilai precision yang dihasilkan menunjukkan bahwa perolehan nilai yang benar-benar positif yang dilakukan oleh model DistilBERT sangat baik. Berdasarkan nilai recall yang dihasilkan, menunjukkan bahwa model banyak menemukan nilai yang terprediksi positif dengan baik. Nilai F1-Score yang tinggi menunjukkan hubungan precision dan recall sangat seimbang. Nilai accuracy menjadi nilai yang berperan penting dalam mengukur kinerja model dalam melakukan tugas. Sehingga dari nilai accuracy tersebut, model melakukan kinerja yang bagus dalam melakukan klasifikasi.
